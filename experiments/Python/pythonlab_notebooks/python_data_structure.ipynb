```python
# List
list_1 = []
list_2 = [1, 2, 3]
list_3 = ['a', 'b']

print(list_1)
print(list_2)
print(list_3)
print('')

# List functions (Part 1)
list_2.append(4)
list_2.insert(0, -1)
list_3.extend(list_2)

print(list_2)
print(list_3)
print('')

# List functions (Part 2)
del list_2[1]
pop = list_2.pop(-1)
list_3.remove(4)

print(list_2)
print(pop)
print(list_3)
print('')

# List access
print(list_3)
print(list_3[0:4])
print(list_3[0:4:2])
print(list_3[::-1])
print('')
```

```python
# Stack - Class
class Stack:
    def __init__(self):
        self.items = []
        
    def __str__(self):
        string = ''
        for item in self.items:
            string += str(item) + ', '
        return string[0:-2]
        
    def is_empty(self):
        return self.items == []
    
    def push(self, data):
        self.items.append(data)
    
    def pop(self):
        return self.items.pop()
```

```python
# Stack
stack_1 = Stack()
stack_2 = Stack()

# Stack functions (Part 1)
stack_2.push(1)
stack_2.push(2)
stack_2.push(3)
stack_2.push(4)
stack_2.push(5)

print(stack_2)
print('')

# Stack functions (Part 2)
stack_2.pop()
stack_2.pop()
stack_2.pop()

print(stack_2)
print('')
```

```python
# Tree & NodeTree - Class
class NodeTree:
    def __init__(self, value):
        self.left = None
        self.right = None
        self.value = value
        
    def _inorder(self):
        items = []
        
        if self.left is not None:
            left = self.left._inorder()
            items.extend(left)
                
        items.append(self.value)
        
        if self.right is not None:
            right = self.right._inorder()
            items.extend(right)
        return items
    
    def _preorder(self):
        items = []
        
        items.append(self.value)
        if self.left is not None:
            left = self.left._preorder()
            items.extend(left)
            
        if self.right is not None:
            right = self.right._preorder()
            items.extend(right)
        
        return items
    
    def _postorder(self):
        items = []
        
        if self.left is not None:
            left = self.left._postorder()
            items.extend(left)
        if self.right is not None:
            right = self.right._postorder()
            items.extend(right)
        items.append(self.value)
        
        return items
    
class Tree:
    def __init__(self):
        self.root = None
        
    def get_root(self):
        return self.root
    
    def add(self, value):
        if self.root is None:
            self.root = NodeTree(value)
        else:
            self._add(value, self.root)
            
    def _add(self, value, node):
        if node.value > value:
            if node.left is None:
                node.left = NodeTree(value)
            else:
                self._add(value, node.left)
        else:
            if node.right is None:
                node.right = NodeTree(value)
            else:
                self._add(value, node.right)
                
    def find(self, value):
        if self.root is not None:
            return self._find(value, self.root)
        
    def _find(self, value, node):
        result = False
        if value == node.value:
            result = True
        elif value < node.value and node.left is not None:
            result = self._find(value, node.left)
        elif value > node.value and node.right is not None:
            result = self._find(value, node.right)
        return result
    
    def print_inorder(self):
        items = self.root._inorder()
        print(items)
        
    def print_preorder(self):
        items = self.root._preorder()
        print(items)
        
    def print_postorder(self):
        items = self.root._postorder()
        print(items)
```

```python
# Tree
tree = Tree()

# Tree functions
tree.add(3)
tree.add(2)
tree.add(1)
tree.add(4)
tree.add(5)

tree.print_inorder()
tree.print_preorder()
tree.print_postorder()
print('')

print(tree.find(2))
print(tree.find(6))
print(tree.get_root().value)
print('')
```

```python
# LinkedList & NodeList - Class
class NodeList: 
    def __init__(self, value=None, next_node=None):
        self.value = value
        self.next = next_node
        
    def set_value(self, value):
        self.value = value
        
    def get_value(self):
        return self.value
    
    def get_next(self):
        return self.next
    
    def set_next(self, next_node):
        self.next = next_node
        
class LinkedList:
    def __init__(self, head=None):
        self.head = head
        
    def __str__(self):
        string = ''
        current = self.head
        while current:
            string = str(current.get_value()) + ', ' + string
            current = current.get_next()
        return string[0:-2]
        
    def insert(self, value):
        node = NodeList()
        node.set_value(value)
        node.set_next(self.head)
        self.head = node
        
    def size(self):
        current = self.head
        size = 0
        while current:
            size += 1
            current = current.get_next()
        return size
    
    def search(self, value):
        current = self.head
        found = False
        while current and found == False:
            if current.get_value() == value:
                found = True
            current = current.get_next()
        return found
    
    def delete(self, value):
        current = self.head
        previous = None
        
        while current:
            if current.get_value() == value:
                if previous is not None:
                    previous.set_next(current.get_next())
                current = None
            else:
                previous = current
                current = current.get_next()
```

```python
# LinkedList
linkedlist = LinkedList()

# LinkedList functions
linkedlist.insert(1)
linkedlist.insert(2)
linkedlist.insert(3)

size = linkedlist.size()
search_1 = linkedlist.search(2)
search_2 = linkedlist.search(-1)
linkedlist.delete(1)
linkedlist.delete(-1)

print(size)
print(search_1)
print(search_2)
print(linkedlist)
print('')
```

```python
from collections import defaultdict

# Graph
graph = defaultdict(list)

def add_edge(graph, key, value):
    graph[key].append(value)
    
def generate_edges(graph):
    edges = []
    for node in graph:
        for neighbour in graph[node]:
            edges.append((node, neighbour))
    return edges

add_edge(graph, 'a', 'c')
add_edge(graph, 'b', 'c')
add_edge(graph, 'b', 'e')
add_edge(graph, 'c', 'd')
add_edge(graph, 'c', 'e')
add_edge(graph, 'c', 'a')
add_edge(graph, 'c', 'b')
add_edge(graph, 'e', 'b')
add_edge(graph, 'd', 'c')
add_edge(graph, 'e', 'c')

print(generate_edges(graph))
```
